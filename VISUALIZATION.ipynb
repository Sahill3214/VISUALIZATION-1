```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate a synthetic dataset with correlated features
np.random.seed(42)
x1 = np.random.randn(100)
x2 = x1 + np.random.randn(100) * 0.5
x3 = x1 + np.random.randn(100) * 0.2

data = {'Feature 1': x1, 'Feature 2': x2, 'Feature 3': x3}
df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Add title
plt.title('Correlation Matrix')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D

# Set random seed
np.random.seed(30)

# Generate data
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], c='blue', marker='o')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D 1  surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set random seed
np.random.seed(25)

# Generate data
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.7)

# Add labels and title
for i, label in enumerate(df['Country']):
    plt.annotate(label, (df['GDP'][i], df['Population'][i]))

plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart: Country Population vs GDP')

# Display the plot
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import pandas as pd
import numpy as np

# Generate random data
np.random.seed(42)
N = 100
x = np.random.rand(N) * 10
y = np.random.rand(N) * 10
sizes = np.random.randint(10, 20, size=N)
colors = ["#"+''.join([np.random.choice(list('ABCDEF0123456789')) for i in range(6)]) for j in range(N)]

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create the plot
output_notebook()
p = figure(title="Scatter Plot with Random Data", x_axis_label='x', y_axis_label='y')

# Add the scatter plot
p.scatter(x='x', y='y', size='sizes', color='colors', source=df)

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create the plot
output_notebook()
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

# Add the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

# Show the plot
show(p)
```
